```python
# %load preamble.py
%env CUDA_VISIBLE_DEVICES=0

import tensorflow as tf
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
session = tf.Session(config=config)

import keras
import keras.backend as K
import keras.layers as layers
from keras.models import Model
from keras.utils import to_categorical
import sklearn.metrics
from keras.callbacks import Callback, EarlyStopping

import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
# super inefficient simulation of a ball with random initial
# velocity that bounces in a box bounded by (0,0),(1,1).
# steps - number of timesteps to simulate
# g     - gravitational constant
# vmax  - maximum inital velicity
# inel  - inelasticity of ball during bounce
def mk_ball(steps=25, g=1./50, vmax=0.25, inel=0.15):
    v0 = np.random.rand()*vmax
    x0 = np.random.rand()
    y0 = np.random.rand()
    theta = np.random.rand()*2*np.pi
    pos = []
    pos.append((x0, y0))
    vx = v0*np.cos(theta)
    vy = v0*np.sin(theta)
    oob_x = False
    oob_y = False
    ndiv = 500
    dt = 1./ndiv
    for istep in range((steps-1)*ndiv):
        x0 = x0 + vx*dt
        y0 = y0 + vy*dt - 0.5*g*(dt**2)
        vy = vy - g*dt
        if not oob_x:
            if x0>1 or x0 < 0:
                oob_x = True
                vx = -vx*(1-inel)
        else:
            if x0 > 0 and x0 < 1:
                oob_x = False
        if not oob_y:
            if y0>1 or y0 < 0:
                oob_y = True
                vy = -vy*(1-inel)
        else:
            if y0 > 0 and y0 < 1:
                oob_y = False
        if istep%ndiv==0:
            pos.append((x0,y0))
    return np.array(pos)
```

```python
# test the generator
b = mk_ball(10)
print(b.shape)
plt.plot(b[:,0], b[:,1],'o-')
plt.ylim(0,1)
plt.xlim(0,1)
```

```python
# number of steps to show the LSTM, and number of points
# for the output to extrapolate
nstep = 11
nextrap = 1

# number of extrapolations to perform for plots
nextend = 10
```

```python
# generate training and validation data.
# the simulator is very dumb, so this is slow...
ntrain, nval = 12000, 4000

train = np.array([mk_ball(nstep+nextrap) for _ in range(ntrain)])
trainX = train[:,:-nextrap]
trainY = train[:,-nextrap:]

val = np.array([mk_ball(nstep+nextrap) for _ in range(nval)])
valX = val[:,:-nextrap]
valY = val[:,-nextrap:]

print(train.shape)
print(trainX.shape)
print(trainY.shape)
```

```python
# simple LSTM+dense network
def mk_model():
    K.clear_session()
    
    def s(L, name=None):
        # activation for dense layers
        return layers.PReLU(name=name)(L)
    
    ball_in = H = layers.Input((nstep,2), name='ball_in')
    H = layers.LSTM(64)(H)
    H = layers.Dense(32)(H)
    H = s(H)
    H = layers.Dense(16)(H)
    H = s(H)
    H = layers.Dense(2*nextrap, activation='sigmoid')(H)
    H = layers.Reshape((nextrap, 2))(H)
    ball_out = H

    model = Model(ball_in, ball_out)
    model.compile(optimizer='rmsprop', loss='mse')
    
    return model
```

```python
# callback class to plot examples at the end of epochs
class PlotCB(Callback):
    # freq = frequency with which to plot outputs (i.e. freq=5 will plot every 5th epoch)
    def __init__(self, freq=None):
        Callback.__init__(self)
        self.freq = freq
        
    def on_epoch_end(self, epoch, logs=None):
        if self.freq and epoch%self.freq: return

        plt.figure()
        plt.suptitle("epoch %d trajectories"%epoch)

        diffs = []
        # plot some example trajectories and the current model's predictions
        for iplot in range(4):
            plt.subplot(2,2,iplot+1)
            truth = mk_ball(nstep+nextrap*nextend)
            seed = truth[:nstep].copy()
            
            record = seed.copy()
            
            plt.plot(truth[:,0], truth[:,1], color='black')
            plt.plot(seed[:,0], seed[:,1],'o', color='C0')
            plt.ylim(0,1)
            plt.xlim(0,1)

            for istep in range(nextend):
                p = self.model.predict(np.expand_dims(seed, axis=0))[0]
                seed = np.vstack([seed,p])[nextrap:]
                
                record = np.vstack([record, p])
            plt.plot(record[:,0], record[:,1], '-o', color='C4', alpha=0.4)
            diffs.append(np.sqrt(np.sum((record-truth)**2,axis=-1)))
            
        # plot residuals between predicted and true ball position
        plt.figure()
        plt.suptitle("epoch %d residuals"%epoch)
        for iplot in range(4):
            plt.subplot(2,2,iplot+1)
            plt.plot(diffs[iplot])
            if iplot == 2:
                plt.xlabel('steps', fontsize=12)
                plt.ylabel(r'$\Delta(x_p - x_0)$', fontsize=12)
```

```python
# (re)initialize the model
model = mk_model()
model.summary();
```

```python
# train the model
model.fit(trainX, trainY,
          validation_data=(valX,valY),
          batch_size=64, epochs=51,
          callbacks=[PlotCB(freq=10)],
         );
```

```python
plt.plot(model.history.history['loss'], label='train')
plt.plot(model.history.history['val_loss'], label='validation')
plt.xlabel('epoch', fontsize=14)
plt.ylabel('MSE', fontsize=14)
plt.yscale('log');
plt.legend();
```

```python
# an alternative approach: train an LSTM to "shift" the 
# input sequence i_0..i_k -> i_1..i_(k+1), where i_(k+1)
# is the extrapolated point
def mk_shifter():
    shift_in = H = layers.Input((None, 2))
    H = layers.LSTM(128, return_sequences=True)(H)
    #H = layers.Dense(128, activation='relu')(H)
    #H = layers.LSTM(128, return_sequences=True)(H)
    H = layers.LSTM(2, return_sequences=True)(H)
    shift_out = H
    
    model = Model(shift_in, shift_out)
    model.compile(optimizer='adam', loss='mse')
    
    return model
```

```python
shifter = mk_shifter()
shifter.summary()
```

```python
shifter.fit(trainX[:,:-1], trainX[:,1:],
            validation_data=(valX[:,:-1],valX[:,1:]),
            batch_size=128, epochs=51,
           )
```
